```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(4,15,100)
y = np.sin(x)
```

```python
plt.figure(figsize=(5,3))
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x,y,color="red")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x,y,"b--")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x,y,"g-.") #k is for black
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x,y,"k--",label="Line") #k is for black
plt.legend() 
plt.show()
```

```python
x =  np.array(['A','B','C','D','E'])
y =  np.array([4, 6, 3, 5, 7])
```

```python
x =  np.random.rand(100)
y =  np.random.rand(100)
colors = np.random.rand(100)
size = 10000 * np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
#you can change alpha which is transpa
plt.scatter(x,y,c=colors,s=size,cmap="plasma",alpha=0.5) 
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.scatter(x,y,c=colors,s=size,cmap="viridis",alpha=0.5) 
plt.colorbar(label="Intensity")
plt.show()
```

```python
data = np.random.rand(1000)
plt.figure(figsize=(5,3))
plt.title("Histogram")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.hist(data,histtype="bar",bins=20,color="yellow",edgecolor="black",alpha=0.9)
plt.show()
```

```python
size = np.array([34, 56, 21,69])
labels = np.array(["Ram","Sham","Sita","Gita"])
```

```python
plt.figure(figsize=(5,4))
plt.pie(size,labels=labels,explode=[0.0,0.0,0.1,0.0],autopct="%1.1f%%",colors=["r", "g", "b", "y"])
plt.title("Pie Chart")
plt.axis("equal")
plt.show()
```

```python
plt.figure(figsize=(5,4))
cols = ["cyan","teal","turquoise","skyblue"]
plt.pie(size,labels=labels,explode=[0.0,0.0,0.1,0.0],autopct="%1.1f%%",colors=cols)
plt.title("Pie Chart")
plt.axis("equal")
plt.show()
```

```python
angle = np.linspace(0,2*np.pi,100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize=(6,5))
ax = plt.subplot(111,polar=True)
ax.plot(angle,radii,color="red",linewidth=2)
ax.fill(angle,radii,color="red",alpha=0.3)
plt.title("Polar Plot")
plt.show
```

```python
overs = np.arange(1,21)
ind = np.random.choice(range(0,25),20)
nz = np.random.choice(range(0,25),20)
```

```python
ind
```

```python
nz
```

```python
plt.figure(figsize=(20,5))
#Line Plot
#1 row 2 coulmn 1 line dikhega
plt.subplot(121)
plt.plot(ind,color="blue",label="IND")
plt.plot(nz,color="black",label="NZ")
plt.legend()

#Bar Chart
plt.subplot(122)
plt.bar(overs+1,ind,color="blue",label="IND")
plt.bar(overs+1,nz,color="black",label="NZ")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
#Line Plot
#1 row 3 column usme se 1
plt.subplot(131)
plt.plot(ind,color="blue",label="IND")
plt.plot(nz,color="black",label="NZ")
plt.legend()

#Bar Chart
#1 row 3 column usme se 2
plt.subplot(132)
plt.bar(overs+1,ind,color="blue",label="IND")
plt.bar(overs+1,nz,color="black",label="NZ")
plt.legend()

#Scatter 
#1 row 3 column usme se 3
plt.subplot(133)
plt.scatter(ind,nz)
plt.show()
```
