```python
net = pd.read_csv("Netflix Userbase.csv")
net.head()
```

```python
from scipy.stats import chi2_contingency
```

```python
sub_type_country = pd.crosstab(net["Subscription Type"],net["Country"])
```

```python
stat_value,p_value,_,_ = chi2_contingency(sub_type_country)
```

```python
sub_type_country
```

```python
stat_value,p_value,_,_ = chi2_contingency([[60,97], [35,54]])
p_value
```

```python
if p_value > 0.05:
    print("Not any dependency")
else:
    print("Columns are  dependent")
```

```python
stat_value,p_value,_,_ = chi2_contingency([[13,2], [4,11]])
p_value
```

```python
if p_value > 0.05:
    print("Not any dependency")
else:
    print("Columns are  dependent")
```
